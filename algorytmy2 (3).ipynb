Zaproponuj rekurencyjny algorytm do znalezienia największego elementu w podanej sekwencji elementów S. Jaka będzie złożoność tego algorytmu? Ile zajmie pamięci?


```python
import time, numpy as np, matplotlib.pyplot as plt, random
from scipy.optimize import curve_fit
def fun_max(S, max=0):
  ''' funkcja znajduje rekurencyjnie największy element w podanej sekwecji
      input : S - lista (sekwencja)
      output : element największy '''
  n=len(S)
  if len(S)==0:
    return max
  else:
    if max>S[0]:
      return fun_max(S[1::],max)
    if max<S[0]:
      max=S[0]
      return fun_max(S[1::],max)
def func(x,a,b):
  return a*x+b
start = time.time()
S=random.sample(list(range(0,2000)), 100)
end = time.time()
#print(S)
print(fun_max(S), end-start)
xs = list(range(0,1000,100))
ys = []
for i in xs:
  start = time.time()
  fun_max(random.sample(list(range(0, 2000)), i))
  end = time.time()
  ys.append(end-start)
popt, pcov = curve_fit(func, xs, ys)
x2 = np.array(xs)
plt.plot(xs,ys,'ro',label="Dane")
plt.plot(x2, func(x2, *popt), label="Hipoteza")
plt.xlabel("Rozmiar")
plt.ylabel("Czas wykonania")
plt.legend(loc='upper left')
plt.title("func_max")
plt.show()
```

Złożoność tego algorytmu to O(n)

Poniższy kod przedstawia implementację rekurencyjnego algorytmu wyliczania n-tej potęgi liczby x:

Korzystając z przykładu podanego na wykładzie, narysuj rekurencyjny
ślad wywołania power(2,5).


```python
def power(x, n):
  """Compute the value x**n for integer n."""
  if n == 0:
    return 1
  else:
    return x*power(x, n-1)
```

Napisz rekurencyjną funkcję w Pythonie, która znajdzie namniejszą i

największą wartość w podanej sekwencji bez używania pętli.


```python
def mm(s, min=2000, max=0):
  ''' funkcja znajduje rekurencyjnie wartość najmniejszą i największą
      input : s - sekwencja
      output : wartość największa i najmniejsza w podanej sekwencji'''
  if len(s)==0:
    return min, max
  else:
    if max<s[0]:
      max=s[0]
    if min>s[0]:
      min=s[0]
    return mm(s[1::],min,max)

import random
S=random.sample(list(range(0,2000)), 100)
print(S)
print(mm(S))
print('funkcje wbudowane: ',min(S),' ',max(S))
```

Napisz rekurencyjną funkcję w Pythonie, która policzy iloczyn dwóch
liczb naturalnych m i n tylko przy użyciu operatorów dodawania i odejmowania.


```python
def dod(m,n,i=1,w=0):
  ''' funkcja liczy iloczyn dwóch liczb za pomocą dodawanie i odejmowania
      input : m,n - liczby naturalne
      output wynik - liczba naturalna '''
  if m>n :
    if i>n:
      return w
    else:
      return dod(m,n,i+1,w+m)
  else:
    if i>m:
      return w
    else:
      return dod(m,n,i+1,w+n)

dod(100,2)
```

Napisz rekurencyjną funkcję w Pythonie, która sprawdzi, czy zadany
łańcuch znaków jest palindromem

```python
def pal(w):
  ''' funkcja sprawdza czy podany łańcuch jest palindromem
      input : w - string, łańcuch znaków, który sprawdzamy
      output : bool - wartość prawda/fałsz'''
  l=len(w)
  if l==1:
    return True
  if l==2:
    return w[0]==w[1]
  else:
    #print("else")
    if w[0]==w[l-1]:
      #print('rek')
      return pal(w[1:l-1])
    else:
      return 1==0

print(pal('kajafk'))
print(pal('kamilślimak'))
```

Napisz rekurencyjną funkcję w Pythonie o sygnaturze
find(path,filename), która startując od zadanej ścieżki znajdzie wszystkie pliki o nazwie podanej jako drugi argument.


```python
import os
def find(path, filename):
    '''funkcja znajduje rekurencyjnie pliki o podanej nazwie startując od zadanej ścieżki
        input : path - ścieżka, od której funkcja szuka plików
                filename - nazwa szukanych plików
        output : lista ścieżek do plików o podanej nazwie'''
    tab = []

    files = os.listdir(path)
    for i in files:
        a = os.path.join(path, i)
        if os.path.isdir(a):
            tab += find(a, filename)
        elif os.path.isfile(a):
            if filename in i:
                tab.append(a)
    return tab

print(find(r'C:\Users\lenovo\Desktop\maria\l2_z6_dir','howareyou.txt'))

```
